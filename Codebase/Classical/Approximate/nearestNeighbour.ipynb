# Nearest-Neighbor Algorithm for TSP (Approximate Solution)

*Time Complexity: O(n^2)*
<br>
*Space Complexity: O(n^2)*

```python
import numpy as np
import time
```

```python
class TSPNearestNeighborSolver:
    def __init__(self, distanceMatrix):
        """
        Initialize the TSP solver with the distance matrix.

        :param distanceMatrix: 2D list containing distances between cities
        """
        self.distanceMatrix = np.array(distanceMatrix)
        self.numCities = len(distanceMatrix)
        self.visitedCities = np.zeros(self.numCities, dtype=bool)
        self.bestRoute = []
        self.minCost = 0

    def solve(self, startCity=0):
        """
        Solves the TSP using the nearest-neighbor algorithm.

        :param startCity: The city to start the tour from (default is 0)
        :return: Tuple containing the best route and the minimum cost
        """
        start_time = time.time()
        self.visitedCities[:] = False
        self.bestRoute = []
        self.minCost = 0

        # Start from the given city
        currentCity = startCity
        self.bestRoute.append(currentCity)
        self.visitedCities[currentCity] = True

        for _ in range(1, self.numCities):
            nextCity = self.findNearestNeighbor(currentCity)
            self.bestRoute.append(nextCity)
            self.minCost += self.distanceMatrix[currentCity, nextCity]
            self.visitedCities[nextCity] = True
            currentCity = nextCity

        # Return to the starting city to complete the cycle
        self.minCost += self.distanceMatrix[currentCity, startCity]
        self.bestRoute.append(startCity)

        end_time = time.time()
        print(f"Runtime: {end_time - start_time:.4f} seconds")
        return self.bestRoute, self.minCost

    def findNearestNeighbor(self, currentCity):
        """
        Find the nearest unvisited city from the current city.

        :param currentCity: Index of the current city
        :return: Index of the nearest unvisited city
        """
        distances = self.distanceMatrix[currentCity]
        unvisited = np.where(~self.visitedCities)[0]
        nearestCity = unvisited[np.argmin(distances[unvisited])]
        return nearestCity
```

```python
class TSPApp:
    def __init__(self):
        """
        Initialize the application.
        """
        self.distanceMatrix = None
        self.solver = None
    
    def setDistanceMatrix(self, distanceMatrix):
        """
        Set the distance matrix and initialize the solver.

        :param distanceMatrix: 2D list containing distances between cities
        """
        self.distanceMatrix = distanceMatrix
        self.solver = TSPNearestNeighborSolver(distanceMatrix)
    
    def run(self, startCity=0):
        """
        Run the TSP solver and display the results.

        :param startCity: The city to start the tour from (default is 0)
        """
        if self.solver is None:
            raise Exception("Distance matrix is not set.")
        
        bestRoute, minCost = self.solver.solve(startCity)
        self.displayResults(bestRoute, minCost)
    
    def displayResults(self, bestRoute, minCost):
        """
        Display the best route and minimum cost.

        :param bestRoute: Best route found
        :param minCost: Minimum cost of the route
        """
        print("Best Route: ", bestRoute)
        print("Minimum Cost: ", minCost)
    
    def generateRandomGraph(self, n, max_distance=100):
        """
        Generate a random distance matrix for n cities.

        :param n: Number of cities
        :param max_distance: Maximum distance between any two cities
        :return: 2D list representing the distance matrix
        """
        distanceMatrix = np.random.randint(1, max_distance + 1, size=(n, n))
        np.fill_diagonal(distanceMatrix, 0)
        
        return distanceMatrix

    def printPath(self, path):
        """
        Print the path in list or array form.

        :param path: List representing the path (e.g., [0, 2, 3, 1, 0])
        """
        path_array = np.array(path)
        print("Path in array form:", path_array)
```

```python
# Example usage
n = 30  # Number of cities
tspApp = TSPApp()
distanceMatrix = tspApp.generateRandomGraph(n)
tspApp.setDistanceMatrix(distanceMatrix)

print("Distance Matrix :\n", distanceMatrix)
```

```python
tspApp.run(startCity=0)
```
