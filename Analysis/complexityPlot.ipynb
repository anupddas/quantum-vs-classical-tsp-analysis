#### Time Complexity

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.table import Table
from math import factorial
```

```python

n_values = np.arange(1, 25)
```

```python
brute_force = [factorial(n) for n in n_values]
bellman_held_karp = [n**2 * 2**n for n in n_values]
greedy = [n**2 * np.log2(n) for n in n_values]
nearest_neighbour = [n**2 for n in n_values]

I = 100     # No of iterations
S = 1024    # Sample Size
p = 1       # QAOA Layer

QAOA = [I*S*p*(n**2) for n in n_values]

quantum_annealing = [100 * (n**3) for n in n_values]
```

```python
# 1) Create figure at high DPI (e.g. 300 for print‑quality)
plt.figure(figsize=(20, 10), dpi=1200)

# heavy, distinctive markers for the three primary algorithms
plt.plot(
    n_values, brute_force,
    label='Brute‑Force',
    linestyle='-',
    linewidth=5,
    marker='o',            # circle
    markersize=10,
    markeredgewidth=2
)
plt.plot(
    n_values, bellman_held_karp,
    label='Bellman‑Held‑Karp',
    linestyle='--',
    linewidth=5,
    marker='s',            # square
    markersize=10,
    markeredgewidth=2
)

# lighter weight, different markers for heuristics
plt.plot(
    n_values, greedy,
    label='Greedy',
    linestyle=':',
    linewidth=2.5,
    marker='D',            # diamond
    markersize=8,
    markeredgewidth=1.5
)
plt.plot(
    n_values, nearest_neighbour,
    label='Nearest Neighbour',
    linestyle='-',
    linewidth=2.5,
    marker='v',            # triangle_down
    markersize=8,
    markeredgewidth=1.5
)
plt.plot(
    n_values, QAOA,
    label='QAOA',
    linestyle='--',
    linewidth=2.5,
    marker='x',            # x
    markersize=8,
    markeredgewidth=2
)
plt.plot(
    n_values, quantum_annealing,
    label='Quantum Annealing',
    linestyle='-.',
    linewidth=5,
    marker='^',            # triangle_up
    markersize=10,
    markeredgewidth=2
)

plt.title("Complexity Comparison of Various Algorithms")
plt.xlabel("n (Problem Size)")
plt.ylabel("Operations (Growth Rate)")
plt.yscale('log')

plt.legend(
    fontsize=16,
    loc='upper left',
    frameon=True,
    framealpha=1,
    fancybox=True,
    shadow=True,
    borderpad=1
)
plt.grid(True, which="both", ls="--")

plt.tight_layout()

# Finally display
plt.show()
```
